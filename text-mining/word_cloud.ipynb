```python
# read in the 'cleaned' data
sentences = pd.read_csv('../data/model_input.csv')
sentences.head()
```

## Remove Stopwprds

Although the sentences have been cleaned, the stopwords have not been removed. We are only interested in display words that are meaningful, therefore we remove the stopwords from the sentences. 

```python
# make stopwords
stop = set(stopwords.words('english'))

# remove stopwords
sentences['text'] = sentences['text'].str.split(' ').apply(lambda x: ' '.join(k for k in x if k not in stop))
```

## Top 50 (cleaned) Word frequencies in the Dataset

The code below produces the top 50 most common words in the entire dataset. The words in the plot give us an indication into the most common words spoken across all speeches. The words below can be compared to the words in each of the president's word cloud, which will give us an indication into the subtle word choice differences used by each president. 

```python
all_words = sentences['text'].str.split(expand=True).unstack().value_counts()
data = [go.Bar(
            x = all_words.index.values[2:50],
            y = all_words.values[2:50],
            marker= dict(colorscale='Jet',
                         color = all_words.values[2:100]
                        ),
            text='Word counts'
    )]

layout = go.Layout(
    title='Top 50 (cleaned) Word frequencies in the Dataset'
)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='basic-bar')
```

## Isolate the Senteneces Spoken by each President

```python
# get the sentences for each user
deKlerk = sentences[sentences['labels']==0]["text"].values
Mandela = sentences[sentences['labels']==1]["text"].values
Mbeki = sentences[sentences['labels']==2]["text"].values
Motlanthe = sentences[sentences['labels']==3]["text"].values
Zuma = sentences[sentences['labels']==4]["text"].values
Ramaphosa = sentences[sentences['labels']==5]["text"].values
```

```python
## Create Word Cloud for Each President

# word cloud for deKlerk
plt.figure(figsize=(16,13))
wc = WordCloud(background_color="black", max_words=20, stopwords=STOPWORDS, max_font_size= 40)
wc.generate(" ".join(deKlerk))
plt.title("deKlerk Word Cloud", fontsize=40)
plt.imshow(wc.recolor( colormap= 'Pastel2' , random_state=17), alpha=0.98)
plt.axis('off')
```
